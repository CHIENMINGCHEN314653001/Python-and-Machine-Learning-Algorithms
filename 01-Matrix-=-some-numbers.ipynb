## Main idea

Let  
$$A = \begin{bmatrix}
 a_{11} & a_{12} & \cdots & a_{1n} \\
 \vdots & \vdots & ~ & \vdots \\
 a_{m1} & a_{m2} & \cdots & a_{mn}
\end{bmatrix}$$
be an $m\times n$ matrix.  Let  
$${\bf v} = \begin{bmatrix}
 c_1 \\
 c_2 \\
 \vdots \\
 c_n
\end{bmatrix}$$
be a vector in $\mathbb{R}^n$.  

Then  
$$(A{\bf v})_i = \sum_{j = 1}^n a_{ij}c_j.$$

##### Exercise 5
Let  
```python
p = np.arange(4)
x = np.arange(1,5)
a = x[:,np.newaxis]**p
b = np.array([4, 15, 40, 85])
```
Find a vector `v` such that `np.dot(a, v)` is `b`.

Hint: Consider the system of linear equations  
$x_0 + x_1 + x_2 + x_3 = 4$  
$x_0 + 2x_1 + 4x_2 + 8x_3 = 15$  
$x_0 + 3x_1 + 9x_2 + 27x_3 = 40$  
$x_0 + 4x_1 + 16x_2 + 64x_3 = 85$

````python
v = np.linalg.solve(a, b)
print(v)
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Load the image
img = plt.imread('/incrediville-front.jpg')
h, w, c = img.shape

# Reshape to a list of pixels
pixels = img.reshape(h * w, 3)
print(pixels.shape)

# Convert to grayscale using a simple average
# pixels_g will have shape (h*w,)
pixels_g = pixels.dot(np.array([1/3, 1/3, 1/3]))

# Reshape back to the original 2D dimensions
img_g = pixels_g.reshape(h, w)

# Display the result
plt.imshow(img_g, cmap="gray")
plt.show()
````
